```python
from input_output import *
import altair as alt
import itertools
print("Available datasets:\n\t" + ',\n\t'.join(os.listdir(sims_root())))
```

```python
dataset_names = ["Allosteric_Pyramid_Full_Fall22", "Allosteric_Pyramid_Partial_Fall22", "NoAlloPyramid_14Oct22"]
### Readability ### 
# make sure the order is the same as dataset_names!!!!
dataset_names_readable = ["Full Allosteric", "Partial Allosteric", "Non-Allosteric"]
group_name = "Pyramid"
target_name = "pyramid"
```

```python
# Edit this cell to specify cutoffs
cutoffs = [0.5, 0.7, 0.9, 1.0]
```

```python
assert(len(dataset_names) > 0)
datasets = [choose_results(dsname) for dsname in dataset_names]
```

```python
datasets_data = []
for ic, cutoff in enumerate(cutoffs):
    for dataset, dsname,dsname_readable in zip(datasets, dataset_names, dataset_names_readable):
        print(f"Analyzing {dsname}")
        data = dataset.get_stats(target_name, cutoff, overreach=False)
        # data = dataset.get_flat_yields(target_name, cutoff, overreach=overreach)
        data['dsname'] = dsname_readable
        data['cutoff'] = cutoff
        datasets_data.append(data)
data = pd.concat(datasets_data, axis='index').reset_index()

shortest_timeseries = min(itertools.chain.from_iterable([
    [r.num_timepoints() 
         for r in dataset.flat_runs()] 
    for dataset in datasets
]))
mintime = shortest_timeseries * OXDNA_PRINT_CLUSTERS_EVERY
data = data[data["time"] < mintime]
```

```python
nt = 0
ymax = 8
temperatures = [0.01, 0.025]

df = data[data["nt"] == nt]
df["yield"] = df["yield"] / 8
df["time"] = df["time"] / 1e6 # convert stesps to megasteps
```

```python
charts = []
x_axis_format = '.0%'
for i, t in enumerate(temperatures):
    charts.append(alt.Chart(df[df['temp'] == t]))
    charts[i] = charts[i].properties(title=f"Temp = {t}")
    charts[i] = charts[i].mark_line().encode(
        x=alt.X('time:Q', title="Time (megasteps)", axis=alt.Axis(tickCount=5)),
        y=alt.Y(
            "yield:Q",
            title="Yield",
            scale=alt.Scale(domain=(0,1)),
            axis=alt.Axis(format=x_axis_format, tickCount=4)
        ),
        color=alt.Color('dsname:O', scale=alt.Scale(scheme="accent"), title="Rule"),
        strokeDash=alt.Stroke('cutoff', title="Cutoff")
    )
chart = alt.hconcat(*charts)
chart.configure_axis(
    labelFontSize=18,
    titleFontSize=20
).configure_title(fontSize=24).configure_legend(
        labelFontSize=18,
    titleFontSize=20
)
```
